## Exercise 1 (Tidying a collaboration)

```python
# Dictionary mapping month abbreviations to numerical format
month_conversion = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
}
```

```python
# Original filename
original_filename = 'oct-08_singapore.txt'

print(original_filename)

def convert_filename(original_filename):
    """
        Function to convert filename to desired format
    """
    original_filename = original_filename.split('.')[0] # Remove file extension
    month, day_city = original_filename.split('-')      # Split by '-'
    day, city = day_city.split('_')                     # Split day and city
    return f"{month_conversion[month]}-{day}-{city}.txt"

print(convert_filename(original_filename))
```

```python
rename_my_file = lambda filename : convert_filename(filename)

print(rename_my_file(original_filename))
```

```python
old_filenames = [
    'oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
    'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
    'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
    'jan-01_tokyo.txt'
]
```

```python
# Iterates through every elemnt in old_filename and renames them, storing into accumulator list
for old_filename in old_filenames:
    new_filename = rename_my_file(old_filename)
    print(f"{old_filename} ----> {new_filename}")
```

```python
import glob

# Adjust the rename_my_file function to check if the filename is already in the desired format
def rename_my_file_updated(old_file_name):
    # Check if the filename is already in the desired format (mm-dd-city.txt)
    if old_file_name[:2].isdigit() and old_file_name[3:5].isdigit():
        return old_file_name
    
    return convert_filename(old_file_name)

# Path to the directory containing the .txt files
txt_files_path = os.path.join(".", "os-collaboration-exercise-data", "*.txt")

# Get a list of all .txt files in the directory
txt_files = glob.glob(txt_files_path)

# Rename all .txt files using the adjusted function
renamed_files = []     # Accumulator

# Iterates through every elemnt in old_filename and renames them, storing into accumulator list
for old_file_path in txt_files:
    old_filename = os.path.basename(old_file_path)
    new_filename = rename_my_file_updated(old_filename)
    
    new_file_path = os.path.join(os.path.dirname(old_file_path), new_filename)
    os.rename(old_file_path, new_file_path)
    renamed_files.append(f"{old_filename} ----> {new_filename}")

print(renamed_files[:])
```

```python
cities = set()      # Accumulator

# Extract the list of cities from the renamed file names
for file_path in glob.glob(txt_files_path): cities.add(file_path.split('-')[-1].split('.')[0].capitalize())

# Create a folder for each city in the list
created_folders = []

for city in cities:
    # Define the path for the new folder
    folder_path = os.path.join(".", "os-collaboration-exercise-data", city)
    
    try:
        # Attempt to create the folder, will pass if it already exists
        os.makedirs(folder_path)
        created_folders.append(folder_path)
    except FileExistsError:
        # If the folder already exists, catch the exception and do nothing
        print("Folder already exists!")
        pass

print(created_folders[:])
```

```python
def move_files():
    """
        Function to move files to their corresponding city folders
    """
    moved_files = []
    
    for city in cities:
        # Define the source directory and the target directory for the city
        source_directory = os.path.join(".", "os-collaboration-exercise-data")
        target_directory = os.path.join(source_directory, city)
        
        # Get a list of all files for the city
        city_files = glob.glob(os.path.join(".", "os-collaboration-exercise-data", f"*{city.lower()}.txt"))
        
        for file_path in city_files:
            # Define the target file path
            target_file_path = os.path.join(target_directory, os.path.basename(file_path))

            # Move the file to the target directory
            shutil.move(file_path, target_file_path)
            # Record the move
            moved_files.append(f"Moved {os.path.basename(file_path)} ----> {target_directory}")
    
    return moved_files

moved_files_list = move_files()

print(moved_files_list)
```
