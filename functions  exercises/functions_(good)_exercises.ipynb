## Exercise 1 (Celsius to Fahrenheit or Kelvin)

I have added `try`-`except` since it was covered in the notes! 

```python
def convert_celsius(temperature_celsius, target_scale = 'Fahrenheit'):
    try:
        # Ensure valid target_scale values using assert
        assert target_scale in ['Fahrenheit', 'Kelvin']
        
        match target_scale:
            case 'Fahrenheit':
                return (temperature_celsius * 9/5) + 32
            case 'Kelvin':
                return temperature_celsius + 273.15
    except AssertionError:
        return "Invalid target_scale. Use 'Fahrenheit' or 'Kelvin'."
    except Exception as e:
        return str(e)

celsius_temperature = 100
result0 = convert_celsius(celsius_temperature, 'unknown')
fahrenheit_result = convert_celsius(celsius_temperature, 'Fahrenheit')
kelvin_result = convert_celsius(celsius_temperature, 'Kelvin')

print(result0)
print(f"{celsius_temperature}°C in fahrenheit is {fahrenheit_result}°F")
print(f"{celsius_temperature}°C in kelvins is {kelvin_result}K")
```

## Exercise 2 (Fahrenheit to Celsius or Kelvin)

```python
def convert_fahrenheit(temperature_fahrenheit, target_scale='Celsius'):
    try:
        # Ensure valid target_scale values using assert
        assert target_scale in ['Celsius', 'Kelvin']
        
        match target_scale:
            case 'Celsius':
                return (temperature_fahrenheit - 32) * 5/9
            case 'Kelvin':
                return (temperature_fahrenheit - 32) * 5/9 + 273.15
    except AssertionError:
        return "Invalid target_scale. Use 'Celsius' or 'Kelvin'."
    except Exception as e:
        return str(e)

fahrenheit_temperature = 212  # Replace with your desired temperature in Fahrenheit
result1 = convert_fahrenheit(fahrenheit_temperature, 'unknown')
celsius_result = convert_fahrenheit(fahrenheit_temperature, 'Celsius')
kelvin_result = convert_fahrenheit(fahrenheit_temperature, 'Kelvin')

print(result1)
print(f"{fahrenheit_temperature}°F in celcius is {celsius_result}°C")
print(f"{fahrenheit_temperature}°F in kelvin is {kelvin_result}K")
```

## Exercise 3 (General Temperature Conversion)

```python
def convert_temperature(temperature, source_scale, target_scale):
    try:
        # Ensure valid target_scale values using assert
        assert source_scale in ['Fahrenheit', 'Celsius', 'Kelvin'] or target_scale in ['Fahrenheit', 'Celsius', 'Kelvin']

        # Convert source temperature to Celsius as an intermediate step
        match source_scale:
            case 'Celsius':
                temperature_celsius = temperature
            case 'Fahrenheit':
                temperature_celsius = convert_fahrenheit(temperature)
            case 'Kelvin':
                temperature_celsius = temperature - 273.15
            case _:
                raise ValueError("Unknown source scale")

        # Convert from Celsius to target scale
        match target_scale:
            case 'Celsius':
                return temperature_celsius
            case 'Fahrenheit':
                return convert_celsius(temperature_celsius)
            case 'Kelvin':
                return temperature_celsius + 273.15
            case _:
                raise ValueError("Unknown target scale")
    except AssertionError:
        return "Invalid scales. Use 'Fahrenheit', 'Celsius' or 'Kelvin'."
    except Exception as e:
        return str(e)

temperature = 100
source_scale = 'Celsius'
target_scale = 'Kelvin'

print(convert_temperature(temperature, "unknown", "Fahrenheit"))
print(convert_temperature(temperature, "Celsius", "Fahrenheit"))
```
