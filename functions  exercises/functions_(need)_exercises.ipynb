## Exercise 1 (Do you know why?)

The `if`-`else` statement in `greeting` uses an implicit `else` statement which returns `f'Hello {name}!'`. Python will skip the indented block under the `if` conditional statement if said conditional statement returns `False`. 

When the parameter `name` is passed a value, the function block code first checks if `name` is equivalent in value and type to `'Batman'`. If it is, it will execute the code within the code block, returning the intended message. Then, the computer will exit the function call as the function call has been executed.

Conversely, if it is not, it will skip the `if` code block and jump to the second `return` statement outside the `if` code block. 

Thus, the code works as intended as it is impossible for a function call to return twice! 

## Exercise 2 (Chubby or not)

```python
def calculate_bmi(weight: float, height: float) -> None:
    bmi = round(weight / (height ** 2), 1)
    
    # BMI category dictionary
    bmi_categories = {
        "Underweight": (0, 18.4),   # Rounded to 1d.p., since less than 18.5, it would be less than or equal to 18.4
        "Normal weight": (18.5, 24.9),
        "Overweight": (25, 29.9),
        "Obesity": (30, float("inf"))
    }

    for category, (lower, upper) in bmi_categories.items():
        if lower <= bmi <= upper:
            return category

    return "UNDETERMINED"

weight = 70
height = 1.75

print(f"Your weight in kg is {weight} and your BMI category is {calculate_bmi(weight, height)}!")
```
