```python
def cast_to_int(val: int, new_type, default = None):    ## Function to correctly handle type casting with try-except
    try:
        return new_type(val)
    except (ValueError, TypeError):
        return default

grade = ''

while grade == "" or (grade < 0) or (100 < grade):
    inputStr = input("Enter the student's score:")
    grade = cast_to_int(inputStr, int, "")
    if grade != "" and (0 <= grade <= 100):    # Breaks out of the loop if a correct input in the range has been given
        break
    print(f"You have entered {inputStr} which is an invalid input. Please enter a numerical grade!")

match grade:                    # Match-case to handle conditional outputs
    case grade if grade >= 70:
        letter_grade = "A"
    case grade if grade >= 50:
        letter_grade = "B"
    case grade if grade >= 35:
        letter_grade = "C"
    case _:
        letter_grade = "Fail"

print(f"The student's letter grade is: {letter_grade}")
```
