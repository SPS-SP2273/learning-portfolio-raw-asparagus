```python
def cast_to_int(val: int, new_type, default = None):    ## Function to correctly handle type casting with try-except
    try:
        return new_type(val)
    except (ValueError, TypeError):
        return default

grade = ''

while True: # Bad practice; cut me some slack ><
    grade = cast_to_int(input("Enter the student's score:"), int, "bad_input")
    if grade != "bad_input" and (0 <= grade <= 100):    # Breaks out of the loop if a correct input in the range has been given
        break
    print("You have entered an invalid input. Please enter a numerical grade!")

match grade:                    # Match-case to handle conditional outputs
    case grade if grade >= 70:
        letter_grade = "A"
    case grade if grade >= 50:
        letter_grade = "B"
    case grade if grade >= 35:
        letter_grade = "C"
    case _:
        letter_grade = "Fail"

print(f"The student's letter grade is: {letter_grade}")
```
