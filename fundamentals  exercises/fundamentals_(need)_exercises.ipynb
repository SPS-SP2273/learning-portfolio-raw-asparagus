# Exercise 3 (In your own words)

|  # | Term         |                                                       Description                                                       |
|:--:|--------------|-------------------------------------------------------------------------------------------------------------------------|
|  1 | Function     | block of code which only runs when called; functions takes in parameters as arguments and may or may not return a value |
|  2 | Arguments    | objects passed to a function (in Python it is pass by reference)                                                        |
|  3 | Comments     | code that doesn't run; added to improve human readability                                                               |
|  4 | Indentations | Python's way of denoting code blocks                                                                                    |
|  5 | Packages     | Python libraries, python files containing constant definitions and function definitions in other files                  |
|  6 | `.`          | dot operator; Python's OOP operator to allow for access of an object's method or attributes                             |
|  7 | `:`          | representing indent blocks in functions and loops; string/array slicing; identifying keys in dictionaries               |
|  8 | `[]`         | empty list                                                                                                              |
|  9 | `[[[]]]`     | list of a list of an empty list                                                                                         |
| 10 | `{}`         | empty dictionary                                                                                                        |
