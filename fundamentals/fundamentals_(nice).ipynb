# 1 If if is not enough

```python
statement = "SP2273 is in Python"

match statement.split()[-1]:
    case "JavaScript":
        print("JavaScript")

    case "Python":
        print("Python")

    case "PHP":
        print("PHP")
    
    case "C":
        print("C")

    case "Java":
        print("Java")

    case _:
        print("Microsoft Word")
```

# 2 Ternary operators or Conditional Statements

```python
nationality = 'French'

greeting = "Bonjour!" if nationality == 'French' else "Hello!"
print(greeting)
print(("Hello!", "Bonjour!")[nationality == 'French'])
print({False: "Bonjour!", True: "Hello!"}[nationality == 'French'])
```
