## Exercise 1 (Make me an odd list)

```python
def square_of_odds(n: int):
    squares = []

    # Loop through numbers 0 to n
    for number in range(n):
        if number % 2 == 0:     # Skips to next iteration if number is even
            continue
        squares.append(number ** 2)

    return squares

print(square_of_odds(10))
```

## Exercise 2 (Make me an another odd list)

```python
square_of_oddC = [x ** 2 for x in range(10) if x % 2]

print(square_of_oddC)
```

## Exercise 3 (Time me!)

```python
# For loop
%timeit square_of_odds(10)
```

```python
# List comprehension
%timeit [x ** 2 for x in range(10) if x % 2]
```

The odd list made using a `for` loop with `range()` and `continue` took 368ns

The odd list made using a list comprehension took 322ns

<b>The odd list made using a <u>list comprehension</u> is faster!</b>

# Exercise 4 (A problem of decay)

```python
initial_quantity, decay_rate = 100, 0.05
target_quantity = 0.5 * initial_quantity

years, current_quantity = 0, initial_quantity       # Accumulator

# Decay simulation loop
while current_quantity > target_quantity:
    years += 1
    current_quantity -= current_quantity * decay_rate

print(f"It takes {years} years for the sample to reduce to half its original amount")
```

## Exercise 5 (Changes in CO<sub>2</sub>)

```python
co2_data = [
    (2000, 369.55), (2001, 371.14), (2002, 373.28),
    (2003, 375.80), (2004, 377.52), (2005, 379.80),
    (2006, 381.90), (2007, 383.79), (2008, 385.60),
    (2009, 387.43), (2010, 389.90), (2011, 391.65),
    (2012, 393.85), (2013, 396.52), (2014, 398.65),
    (2015, 400.83), (2016, 404.24), (2017, 406.55),
    (2018, 408.52), (2019, 411.44), (2020, 414.24)
]

significant_increases = []  # Accumulator
previous_co2_level = co2_data[0][1]

# Iterates through CO2 data to find years with significant increases in CO2
for i in range(1, len(co2_data)):
    year, co2_level = co2_data[i]
    change = co2_level - previous_co2_level
    previous_co2_level = co2_level
    if change >= 3:
        significant_increases.append((year, change))

for year, co2_level_change in significant_increases:
    print(f"In {year}, CO2 levels increased by {co2_level_change:.3f} ppm.")    # Updated to print to 3d.p. as suggested for improved readability!
```
