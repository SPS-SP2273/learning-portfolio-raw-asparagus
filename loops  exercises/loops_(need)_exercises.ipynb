## Exercise 1 (Celcius to Farenheit)

```python
temperatures_celsius = [
    0, 5, 10, 15, 20, 25,
    30, 35, 40, 45, 50
]

cel_to_far = lambda c: (c * 9/5) + 32   # Lambda function to convert celcius to farenheit
temp_far = list(map(cel_to_far, temperatures_celsius))

# Print both Celsius and Fahrenheit temperatures
for cel, far in zip(temperatures_celsius, temp_far):
    print(f"{cel}°C converted to Farenheit is {far}°F")
```

## Exercise 2 (Multiplication table)

```python
def print_mTable(m, n):
    padding = len(str(m * n))
    
    # Loops through 1 to m (rows)
    for i in range(1, m + 1):
        print(f"{i} : ", end="")
        # Loops through 1 to n (columns)
        for j in range(1, n + 1):
            product = i * j
            # print(f"{product:<{padding + 1}}", end="")  # Padding to ensure that each column is left justified
            print(product, end="\t")  # Fixed as per recommendation!
        print()

print_mTable(5, 5)
```
