## 4.1.1 for with a list

`for` goes through every element in an iterator!

```python
for name in names:
    print(f"{name} is in the SPS committee!")
```

`name` here is a dummy variable! It can be replaced with anything!

```python
for v in names:
    print(f"{v} is in the SPS committee!")
```

## 4.1.2 for with enumerate

`enumerate` is a function that creates an `enumerate` object!

```python
print(enumerate(["one", "two", "three"], 1))
```

```python
print(list(enumerate(["one", "two", "three"], 1)))
```

which is effectively a `tuple` of `tuple`s!

If we have 2 lists and we don't want to zip it into a `dict`, we can use enumerate to obtain `index` `value` 'pairs'!

```python
majors = ["Life Sciences", "Math", "Math", "Life Sciences", "Life Sciences", "Chemistry", "Physics", "Life Sciences", "Life Sciences"]

for index, name in enumerate(names):
    print(f'{name} is in the SPS committee and has index {index}!')
```

Or we can use it to number print statements as such!

```python
majors = ["Life Sciences", "Math", "Math", "Life Sciences", "Life Sciences", "Chemistry", "Physics", "Life Sciences", "Life Sciences"]

for index, name in enumerate(names):
    print(f'{index + 1}: {name} is in the SPS committee!')
```

Or alternatively, use the `index` to iterate through another list B while iterating through list A!

```python
for index, name in enumerate(names):
    major = majors[index]
    print(f'{name} is a {major} major!')
```

Or, if we want to change the index iterated through by `enumerate`, we can instead change the starting index by passing an additional argument to `enumerate()`!

```python
for index, name in enumerate(names, 1):
    print(f'{index}: {name} is in the SPS committee!')
```

## 4.1.3 for with range

`range()` returns an iterable of datatype `range`

```python
print(range(5))
```

which is an iterable containing values!

```python
for i in range(5):
    print(i)
```

we can specify a range

```python
for i in range(215, 220):
    print(i)
```

and a range and step size!

```python
for i in range(1123, 1234, 13):
    print(i)
```

and we can also use the `len` function to create a `range` of the indices of a list to let us iterate through it using a `for` loop!

```python
for i in range(len(names)):
    name = names[i]
    major = majors[i]        
    print(f"{name} is in the SPS committee and is a {major} major!")
```

Additionally, I could do this as well!

```python
for _ in range(10):
    print("Hi! My name is Jun Rui!")
```

# 4.2 while

Another iterator that Python provides that is similar to `for` is the `while` loop! 

```python
number = 0

while number < 5:
    print(number)
    number += 1
```

# Additional notes (from myself!)

Why are there for and while loops, while not just have only one iterating branching syntax?
 - <i>While loops are used when the number of iterations is not known beforehand, relying on a looping conditional for continuation.</i>
 - <i>For loops are preferred when the number of iterations is known or when iterating over a sequence.</i>

 ## Example!
 
 Linear search -- `while` loop is preferred since we don't know the position or how much of the array we must search through if the list is unsorted!

```python
def linear_search(arr, target):
    index = 0
    while index < len(arr):
        if arr[index] == target:
            return index
        index += 1
    return -1

arr = [5, 3, 8, 6, 1]
target = 6
print(linear_search(arr, target))
```

Binary search -- `for` loop is preferred since number of iterations is determined by array size and we can handle iterating over a range more easily!

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

arr = [1, 3, 5, 7, 9]
target = 7
print(binary_search(arr, target))
```
