# 7.1 A simple plot

## 7.1.1 Letâ€™s look at some code.

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
# plt.grid(alpha=.25)
plt.grid(alpha=.8)
# plt.legend(loc='upper left')
plt.legend(loc='lower left')

plt.show()
```

```python
plt.plot(x, y_2, c='red', ls='dashed', lw=2)

plt.show()
```

## 7.1.2 Adding another plot

We can have multiple plots on the same plot!

```python
# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed')
plt.plot(x, y_2, label='Y2 values', color='blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')

plt.show()
```

## 7.1.3 Yet another plot but with error bars

We can include error bars to a specific plot using `errorbar()`

An Example using constant error values:

```python
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')

plt.show()
```

# 7.2 Better with NumPy

`matplotlib` works natively with `numpy` `array` as well!

```python
import numpy as np
```

```python
np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')

plt.show()
```

## 7.2.1 Adding mathematical functions

Here is a prime example of why `numpy` `array` can be a lot more powerful than native primitive `list`

```python
x = np.linspace(0, np.pi, 1000)

x2 = x**2                  # The math stuff
cos_x = np.cos(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, cos_x, label='cos(x)',
         color='blue')
plt.legend()

plt.show()
```

## We need another axis!

`twinx()`

```python
plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()

plt.plot(x, cos_x, label='cos(x)',color='blue', )
plt.legend(loc='lower right')

plt.show()
```

# 7.3 Saving to disc

`matplotlib` provides a very straightforward way to save your plots to a file using the `savefig()` function, allowing for tons of flexibility in terms of file format, resolution, and save location.

Again, we can use `os.path` to make it OS safe and construct OS specific filepaths

```python
import os

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, cos_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

filename = 'simple-plot.png'
desktop_path = os.path.join(".", filename)
plt.savefig(desktop_path, dpi=150)
```
