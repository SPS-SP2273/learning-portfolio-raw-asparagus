# Exercise 1 (Total recall)

| Term       | Brief description |
|------------|-------------------|
| Subsetting | Subsetting entails extracting specific elements or segments from a data structure (such as arrays, lists, or pandas DataFrames) based on precise criteria or indices. This enables the selection of data from larger datasets.                  |
| Indexing   | Indexing allows for the access and manipulation of individual elements or collections of elements within a data structure through their indices (integer-based) or immutable keys (dictionaries). Python supports both zero-based indexing and negative indexing.                  |
| Slicing    | Subsetting a subsequence from a sequence (data structure) by specifying start index, stop index and step delimited by colons                  |
| Masking    | Filtering data through boolean conditions or against a boolean array to subset a subset of data from larger datasets                  |

# Exercise 2 (Show me the ‘odd’ letters)

```python
import numpy as np

np_array_2d = np.array([[1, "A"], [3, "C"], [2, "B"], [4, "D"],
                        [5, "E"], [7, "G"], [6, "F"], [8, "H"],
                        [10, "J"], [9, "I"]])

odd_letters = lambda arr : arr[arr[:, 0].astype(int) % 2 == 1, 1]

print(odd_letters(np_array_2d))
```
