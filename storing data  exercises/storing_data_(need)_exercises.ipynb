# Exercise 2 (Indexing)

```python
py_list = ["a1", "b2", "c3", "d4", "e5", "f6", "g7", "h8", "i9", "j10"]

# Filters elements containing an even number
odd_only = [e for e in py_list if int(''.join(filter(str.isdigit, e))) % 2]     # Safe typecasting

for x in odd_only:
    print(x)
```

# Exercise 3 (Index Again)

```python
elements = ['Hydrogen',
            'Helium', 'Lithium',
            'Beryllium', 'Boron', 'Carbon',
            'Nitrogen', 'Oxygen',
            'Fluorine',
            'Neon']

print(elements[4])                         # Index 4, i.e. 5th element in the list
print(elements[0 - len(elements) + 4])
```

# Exercise 4 (How many ones)

```python
# I treat 'use the concepts as only using things covered in this chapter

import numpy as np

# number of 1's as in the number of integer 1

numbers = [45, 60, 1, 30, 96, 1, 96, 57, 16, 1,
        99, 62, 86, 43, 42, 60, 59, 1, 1, 35,
        83, 47, 34, 28, 68, 23, 22, 92, 1, 79,
        1, 29, 94, 72, 46, 47, 1, 74, 32, 20,
        8, 37, 35, 1, 89, 29, 86, 19, 43, 61]

numbers_arr = np.array(numbers)

print(np.sum(numbers_arr == 1))

# number of 1's as in the number of integers containing a decimal 1

print(np.sum(np.vectorize(lambda x: '1' in str(x))(numbers_arr)))   # Unsafe typecasting, assuming the
                                                                    # entire list to contain only integers
```
