## Quadratic Equations

### Introduction
<img style="display: block; margin-left: auto; margin-right: auto;" src="https://upload.wikimedia.org/wikipedia/commons/thumb/f/f8/Polynomialdeg2.svg/1024px-Polynomialdeg2.svg.png" width="25%">
<i>(Image from the <u><a href="https://en.wikipedia.org/wiki/Quadratic_equation">Wikipedia page on Quadratic equations</a></u>)</i>

The general form of a quadratic equation is:

$$
ax^2 + bx + c = 0
$$

### Solutions
Provided $a \neq 0$, we can use an elementary algebraic method called <u><a href="https://en.wikipedia.org/wiki/Completing_the_square">completing the square</a></u> to show that a quadratic equation has the following solution:

$$
x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}
$$

### Discriminant
The quantity $\Delta = b^2 - 4ac$ is called the <b>discriminant</b> of the equation and decides the nature of its solutions. The table below shows the various possibilities.
|      **Discriminant**     |                 **Roots**                 |
|:------------------------:|:------------------------------------------|
| $\Delta = b^2 - 4ac = 0$ | A single solution of $-b/\left(2a\right)$ |
| $\Delta = b^2 - 4ac > 0$ | Two distinct solutions                    |
| $\Delta = b^2 - 4ac < 0$ | No real solutions; both are complex.      |

1. What is a Git <b>repository</b>?

<font color="blue">A Git repository is a fundamental component of version control, and a structured collection of files along with their revision history. This facilitates it to serve as a centralized hub for storing, tracking, and managing changes to a project's codebase. A Git repository hosts snapshots of the project at various points in time, enabling developers to navigate through previous versions of the code which is imperative to track and understand the evolution of a project. Git repositories uniquely facilitate collaborative development of projects, allowing multiple developers to work on different aspects of a project simultaneously without having to worry about overwriting each other's modifications or actively managing and tracking code changes. Git repositories also provide mechanisms for branching and merging, enhancing the development workflow and aiding in maintaining the integrity and continuity of the software development process.</font>

2. Git keeps tracks of <font color="blue"><b><u>changes</u></b></font> in the files in our repository.
3. What is a <b>commit</b>?

<font color="blue">A commit represents a snapshot of a Git repository's state at a given point in time. Each commit stores information about specific changes made and attributed to specific users since the previous commit to the branch. Every commit can be identified by a unique hash code, timestamp and author.</font>

4. What is a <b>push</b>?

<font color="blue">A push refers to uploading a local repository's content into a remote repository. A push 'pushes' and synchronizes a local repository and a remote repository to the latest version of the project by transferring the commits from a local repository to a remote repository. Pushing starts with a `git add .` command to add all changes you have made into the local git repository, followed with a `git commit -m "New commit"` to commit staged changes with a message before `git push origin main` which push the commit onto Git!</font>

5. What is a <b>pull</b>?

<font color="blue">A pull refers to fetching and integrating changes from a remote repository's content into a local repository. Pulling starts with a `git fetch` command to download contents from the remote repository before `git merge` which merges the commits and files from the remote repository into a local repository.</font>

6. What is the difference between <b>GitHub</b> and <b>GitHub Desktop</b>?

<font color="blue">GitHub is a web-based hosting service for Git repositories providing a platform for source code management and collaborative development. GitHub enables management features such as issue tracking, pull requests, wikis, and a web interface for managing Git repositories. GitHub is widely used for open-source projects, private repositories, as a collaborative platform for developers worldwide and </font><font color="red"><b>has been owned by Microsoft since 2018</b></font><font color="blue">.

GitHub Desktop is a desktop application massively simplifying the use of Git and GitHub on your local machine. It provides a graphical user interface (GUI) to manage Git repositories, making it more accessible for those who prefer not to use command-line tools. GitHub Desktop facilitates and abstracts common Git tasks such as committing, branching, and pushing changes to GitHub. In my experience, I very much prefer the GUI for manageable merge conflicts.</font>
