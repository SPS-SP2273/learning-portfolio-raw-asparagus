# 1.4 Some tips and tricks

## 1.4.1 Keyboard Shortcuts
Here are some shortcuts for Jupyter Notebooks on Visual Studio Code:

<b>General Shortcuts</b>
- Open Command Palette: `Ctrl` + `Shift` + `p` (Windows/Linux), `command` + `shift` + `p` (macOS)
- Save Notebook: `Ctrl` + `s` (Windows/Linux), `command` + `s` (macOS)
- Toggle Sidebar: `Ctrl` + `b` (Windows/Linux), `command` + `b` (macOS)

<b>Editing and Navigating Cells</b>
- Insert Cell Above: `a` (when a cell is selected)
- Insert Cell Below: `b` (when a cell is selected)
- Delete Cell: `d`, `d` (press `d` twice when a cell is selected)
- Move Cell Up: `Ctrl` + `Arrow Up` (Windows/Linux), `command` + `Arrow Up` (macOS)
- Move Cell Down: `Ctrl` + `Arrow Down` (Windows/Linux), `command` + `Arrow Down` (macOS)
- Edit Cell: `Enter`/`return` (when a cell is selected)
- Run Cell: `Ctrl` + `Enter` (Windows/Linux), `command` + `return` (macOS)
- Run Cell and Select Below: `Shift`/`shift` + `Enter`/`return`
- Run Cell and Insert Below: `Alt`/`option` + `Enter`/`return`
- Toggle Cell Output: `Ctrl` + `Shift` + `Enter` (Windows/Linux), `command` + `shift` + `return` (macOS)

<b>Cell Type Conversion</b>
- Convert to Code Cell: `y` (when a cell is selected)
- Convert to Markdown Cell: `m` (when a cell is selected)

<b>Miscellaneous</b>
- Toggle Line Numbers: `l` (when a cell is in edit mode)
- Undo Cell Operation: `z` (in command mode)
- Redo Cell Operation: `Ctrl` + `y` (Windows/Linux), `command` + `shift` + `z` (macOS) in command mode
- Search and Replace: `Ctrl` + `f` (Windows/Linux), `command` + `f` (macOS) in edit mode
