## 1.3.2 Some basic syntax

The following table shows basic Markdown syntax.

| **Style**           |     **Syntax**     |
|---------------------|:------------------:|
| Bold                | `** **` or `__ __` |
| Italic              |   `* *` or `_ _`   |
| All bold and italic |      `*** ***`     |
| Subscript           |   `<sub> </sub>`   |
| Superscript         |   `<sup> </sup>`   |

e.g. The quick *brown fox* jumps **over the**<sup>lazy</sup><sub>dog</sub>
